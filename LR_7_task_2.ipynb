```python
# Імпорт необхідних бібліотек
from sklearn.datasets import load_iris        # датасет Iris
from sklearn.cluster import KMeans            # алгоритм K-means
import matplotlib.pyplot as plt               # для графіків
import numpy as np

# Завантаження датасету Iris
iris = load_iris()

# Ознаки (довжина/ширина чашолистка і пелюстки)
X = iris.data

# Створення моделі K-means (3 кластери)
kmeans = KMeans(n_clusters=3)

# Навчання моделі на даних
kmeans.fit(X)

# Прогнозування (визначення кластерів для кожної точки)
y_kmeans = kmeans.predict(X)

# Побудова графіка (беремо перші 2 ознаки для відображення)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis')

# Отримання центрів кластерів
centers = kmeans.cluster_centers_

# Відображення центрів кластерів
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200)

# Заголовок графіка
plt.title("Кластеризація набору даних Iris")

# Відображення графіка
plt.show()
```
